### Paso 1: Preprocesamiento
Variables num√©ricas ‚Üí Imputer, Scaler, Transformer
Variables categ√≥ricas ‚Üí Imputer, Encoder

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, silhouette_score, accuracy_score
```

```python
# Cargar el dataset
file_path = "../Data/AB_NYC_2019.csv"
df = pd.read_csv(file_path)

# Eliminar columnas irrelevantes
df.drop(columns=['id', 'host_id', 'host_name'], inplace=True)

# Descomponer la columna de fecha en a√±o, mes y d√≠a
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df['review_year'] = df['last_review'].dt.year
df['review_month'] = df['last_review'].dt.month
df['review_day'] = df['last_review'].dt.day
df.drop(columns=['last_review'], inplace=True)

# Separar columnas num√©ricas y categ√≥ricas
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Definir pipeline para variables num√©ricas
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Definir pipeline para variables categ√≥ricas
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Aplicar transformaciones con ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

# Aplicar el preprocesamiento al dataset
df_preprocessed = preprocessor.fit_transform(df)

# Mostrar las dimensiones del dataset despu√©s del preprocesamiento
df_preprocessed.shape
```

```python
# Corregir el par√°metro de OneHotEncoder (usar sparse=False en su lugar)
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Aplicar transformaciones con ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

# Aplicar el preprocesamiento al dataset
df_preprocessed = preprocessor.fit_transform(df)

# Mostrar las dimensiones del dataset despu√©s del preprocesamiento
df_preprocessed.shape
```

Parece que el problema surge por el OneHotEncoder, ya que genera un n√∫mero extremadamente grande de columnas (m√°s de 48,000). Esto ocurre porque alguna de las columnas categ√≥ricas tiene demasiados valores √∫nicos (probablemente name o neighbourhood_group).

Voy a reducir la dimensionalidad limitando el n√∫mero de categor√≠as √∫nicas y reintentarlo. 

```python
# Limitar el n√∫mero de categor√≠as √∫nicas en columnas categ√≥ricas con alto cardinalidad
high_cardinality_cols = ['name', 'neighbourhood', 'neighbourhood_group']

# Mantener solo las 20 categor√≠as m√°s frecuentes y asignar 'Other' a las menos frecuentes
for col in high_cardinality_cols:
    top_categories = df[col].value_counts().nlargest(20).index  # Obtener las 20 categor√≠as m√°s comunes
    df[col] = df[col].apply(lambda x: x if x in top_categories else 'Other')

# Recalcular columnas categ√≥ricas despu√©s de la reducci√≥n de cardinalidad
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Definir pipeline actualizado para variables categ√≥ricas
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Aplicar transformaciones con ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

# Aplicar el preprocesamiento al dataset
df_preprocessed = preprocessor.fit_transform(df)

# Mostrar las dimensiones del dataset despu√©s del preprocesamiento
df_preprocessed.shape
```

Hemos reducido el n√∫mero de columnas generadas por OneHotEncoder, limitando las categor√≠as en name, neighbourhood y neighbourhood_group a las 20 m√°s comunes y agrupando el resto en "Other".

El dataset preprocesado ahora tiene 61 columnas en lugar de miles.

## Siguiente paso: Clustering y An√°lisis de Siluetas
Voy a realizar clustering con KMeans y evaluar la calidad del clustering con la m√©trica de silueta.

```python
from sklearn.cluster import KMeans

# Aplicar KMeans con un n√∫mero √≥ptimo de clusters (probamos con 4 inicialmente)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_preprocessed)

# Agregar la columna de clusters al DataFrame original
df['cluster'] = clusters

# Evaluar la calidad del clustering con la m√©trica de silueta
silhouette_avg = silhouette_score(df_preprocessed, clusters)
silhouette_avg
```

debido al gran tama√±o del dataset procesado, voy a:

Reducir la dimensionalidad usando PCA antes de calcular la silueta.
Probar con una muestra del dataset en lugar de utilizar el conjunto completo

```python
# Reducir la dimensionalidad con PCA antes del clustering
pca = PCA(n_components=10, random_state=42)
df_pca = pca.fit_transform(df_preprocessed)

# Aplicar KMeans despu√©s de PCA
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_pca)

# Agregar la columna de clusters al DataFrame original
df['cluster'] = clusters

# Calcular la m√©trica de silueta con una muestra del dataset para evitar problemas de memoria
sample_size = min(5000, df_pca.shape[0])  # Usar un m√°ximo de 5000 muestras
sample_indices = np.random.choice(df_pca.shape[0], size=sample_size, replace=False)
silhouette_avg = silhouette_score(df_pca[sample_indices], clusters[sample_indices])

silhouette_avg
```

Ahora, vamos a visualizar el clustering con un scatterplot coloreado por los clusters.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear scatterplot de los primeros dos componentes principales de PCA coloreados por cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=df['cluster'], palette='tab10', alpha=0.7)
plt.title("Visualizaci√≥n de Clusters con PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
